```python
# Load necessary libraries
import pandas as pd

# Path to Excel file (update this path based on where you store the file in Drive)
file_path = '/content/drive/MyDrive/Colab Notebooks/Social Media Analytics/Case Study Analysis-1(XOXO)/xoxoday data.xlsx'

# Load specific sheets
tweets_data = pd.read_excel(file_path, sheet_name='Tweets Data')
retweets_data = pd.read_excel(file_path, sheet_name='Retweets Data ')
user_data = pd.read_excel(file_path, sheet_name='Twitter User Data')

# Display first few rows of each for verification
print("Tweets Data Sample:")
display(tweets_data.head())

print("Retweets Data Sample:")
display(retweets_data.head())

print("User Data Sample:")
display(user_data.head())
```

```python
# Rename columns in Tweets Data
tweets_data = tweets_data.rename(columns={
    "Unnamed: 0": "Source_User",
    "Unnamed: 1": "Target_User",
    "Unnamed: 16": "Tweet_Text",
    "Unnamed: 19": "Hashtags",
    "Unnamed: 20": "Tweet_Date"
})

# Retain only relevant columns
tweets_data = tweets_data[['Source_User', 'Target_User', 'Tweet_Text', 'Hashtags', 'Tweet_Date']]
```

```python
# Reload Retweets Data with the first row skipped
retweets_data = pd.read_excel(file_path, sheet_name='Retweets Data ', skiprows=1)

# Rename relevant columns
retweets_data = retweets_data.rename(columns={
    "Vertex 1": "Tweet_ID",
    "Vertex 2": "Original_User",
    "Colour": "Retweeter",  # this is the 'Retweeter' column as per structure
    # Retweet_Date and Hashtags do not seem to exist — ignore for now
})

# Keep only relevant columns
retweets_data = retweets_data[['Tweet_ID', 'Original_User', 'Retweeter']]
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
import re
```

```python
# 1. Number of Tweets
num_tweets = len(tweets_data)
print(f"Total Tweets: {num_tweets}")
```

```python
# 2. Number of Unique Users
unique_users = tweets_data['Source_User'].nunique()
print(f"Number of Unique Users: {unique_users}")
```

```python
# 3. Most Active Users (Top 10)
most_active_users = tweets_data['Source_User'].value_counts().head(10)
print("\nTop 10 Most Active Users:")
print(most_active_users)
```

```python
# 4. Most Used Hashtags
def extract_hashtags(text):
    if pd.isna(text):
        return []
    return re.findall(r"#(\w+)", text)

tweets_data['Hashtag_List'] = tweets_data['Hashtags'].apply(extract_hashtags)
all_hashtags = [hashtag for sublist in tweets_data['Hashtag_List'] for hashtag in sublist]
hashtag_counts = Counter(all_hashtags)
top_hashtags = hashtag_counts.most_common(10)

print("\nTop 10 Most Used Hashtags:")
for tag, count in top_hashtags:
    print(f"{tag}: {count}")
```

```python
# 5. Time-based Trends (Tweet Frequency Over Time)
tweets_data['Tweet_Date'] = pd.to_datetime(tweets_data['Tweet_Date'], errors='coerce')
tweets_data['Date_Only'] = tweets_data['Tweet_Date'].dt.date

tweets_per_day = tweets_data.groupby('Date_Only').size()

plt.figure(figsize=(12,6))
sns.lineplot(x=tweets_per_day.index, y=tweets_per_day.values, marker='o')
plt.title("Tweet Frequency Over Time")
plt.xlabel("Date")
plt.ylabel("Number of Tweets")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
```

```python
# Download VADER lexicon (only needed once)
nltk.download('vader_lexicon')

# Initialize VADER
sia = SentimentIntensityAnalyzer()

# Apply sentiment scoring to non-null tweets
tweets_data['Sentiment'] = tweets_data['Tweet_Text'].dropna().apply(lambda text: sia.polarity_scores(text)['compound'])

# Classify based on compound score
def classify_sentiment(score):
    if score > 0.05:
        return 'Positive'
    elif score < -0.05:
        return 'Negative'
    else:
        return 'Neutral'

tweets_data['Sentiment_Category'] = tweets_data['Sentiment'].apply(classify_sentiment)

# Show sentiment counts
print("\nSentiment Distribution:")
print(tweets_data['Sentiment_Category'].value_counts())

# Plot sentiment distribution
sns.countplot(data=tweets_data, x='Sentiment_Category', palette='Set2')
plt.title("Tweet Sentiment Distribution")
plt.xlabel("Sentiment Category")
plt.ylabel("Number of Tweets")
plt.show()
```

```python
# Time-based sentiment trend
sentiment_over_time = tweets_data.groupby(['Date_Only', 'Sentiment_Category']).size().unstack().fillna(0)

# Line plot of sentiment trends
sentiment_over_time.plot(kind='line', figsize=(12, 6), marker='o')
plt.title("Sentiment Trends Over Time")
plt.xlabel("Date")
plt.ylabel("Number of Tweets")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Read the uploaded file
nrc = pd.read_csv('NRC-Emotion-Lexicon-Wordlevel-v0.92.txt', sep='\t', names=['Word', 'Emotion', 'Association'])

# Filter only positive associations
nrc = nrc[nrc['Association'] == 1]
```

```python
from collections import defaultdict, Counter
import matplotlib.pyplot as plt

# Step 1: Build dictionary {word: [emotion1, emotion2, ...]}
emotion_dict = defaultdict(list)
for _, row in nrc.iterrows():
    emotion_dict[row['Word']].append(row['Emotion'])

# Step 2: Function to get emotions from tweet text
def get_emotions(text):
    if pd.isna(text):
        return []
    words = text.lower().split()
    emotions = []
    for word in words:
        emotions.extend(emotion_dict.get(word, []))
    return emotions

# Step 3: Apply to tweet text
tweets_data['Emotions'] = tweets_data['Tweet_Text'].apply(get_emotions)

# Step 4: Flatten all emotions and count frequency
all_emotions = [emotion for emotions in tweets_data['Emotions'] for emotion in emotions]
emotion_counts = Counter(all_emotions)

# Step 5: Display dominant emotions
print("Top Dominant Emotions in Tweets:")
for emotion, count in emotion_counts.most_common():
    print(f"{emotion.capitalize()}: {count}")

# Step 6: Plot emotion distribution
plt.figure(figsize=(10, 6))
plt.bar(emotion_counts.keys(), emotion_counts.values(), color='lightcoral')
plt.title("Dominant Emotions in Tweets (NRC Emotion Lexicon)")
plt.xlabel("Emotion")
plt.ylabel("Frequency")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
##Topic Modelling USing LDA
```

```python
import pandas as pd
import gensim
from gensim import corpora
from sklearn.feature_extraction.text import CountVectorizer
from nltk.corpus import stopwords
import nltk
import re

# Download NLTK stopwords if not already done
nltk.download('stopwords')
stop_words = stopwords.words('english')

# Step 1: Preprocess tweets (clean, tokenize, remove stopwords)
def preprocess(text):
    if pd.isna(text):
        return []
    # Remove mentions, hashtags, links, and non-alphabetic characters
    text = re.sub(r"(?:@|#)\w+|http\S+|[^a-zA-Z\s]", "", text.lower())
    tokens = text.split()
    return [word for word in tokens if word not in stop_words and len(word) > 2]

# Apply preprocessing
tweets_data['Clean_Tokens'] = tweets_data['Tweet_Text'].apply(preprocess)

# Step 2: Create Dictionary and Corpus for Gensim
dictionary = corpora.Dictionary(tweets_data['Clean_Tokens'])
corpus = [dictionary.doc2bow(text) for text in tweets_data['Clean_Tokens']]

# Step 3: Build LDA Model
lda_model = gensim.models.LdaModel(
    corpus=corpus,
    id2word=dictionary,
    num_topics=5,               # Choose number of topics
    random_state=42,
    update_every=1,
    chunksize=100,
    passes=10,
    alpha='auto',
    per_word_topics=True
)

# Step 4: Print Topics
print("\nTop Keywords per Topic:")
for i, topic in lda_model.print_topics(num_words=10):
    print(f"Topic {i+1}: {topic}")
```
